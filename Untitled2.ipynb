```python
from collections import deque
def bfs(g, s):
  q=deque([s])
  v=set()
  v.add(s)
  while q:
    n=q.popleft()
    print(n,end=" ")
    for i in g[n]:
      if i not in v:
        v.add(i)
        q.append(i)
g={'A':['B','C'],
   'B':['D','E'],
   'C':['F'],
   'D':[],
   'E':[],
   'F':[]
  }
bfs(g,'A')
```

```python
class Graph:
    def __init__(self):
        self.graph = {}
    def add_edge(self, u, v):
        if u not in self.graph:
            self.graph[u] = []
        self.graph[u].append(v)

    def dfs(self, start, visited=None):
        if visited is None:
            visited = set()
        visited.add(start)
        print(start, end=' ')
        for neighbor in self.graph.get(start, []):
            if neighbor not in visited:
                self.dfs(neighbor, visited)

g = Graph()
g.add_edge(1, 2)
g.add_edge(1, 3)
g.add_edge(2, 4)
g.add_edge(3, 5)

print("Depth First Search starting from vertex 1:")
g.dfs(1)
```

```python
class BankAccount:
    def __init__(self, initial_balance):
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: {amount}")
        else:
            print("Invalid withdrawal amount.")

    def get_balance(self):
        return self.__balance


account = BankAccount(1000)
account.deposit(500)
account.withdraw(200)
print(f"Current Balance: {account.get_balance()}")
```

```python
  class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v):
        if u not in self.graph:
            self.graph[u] = []
        self.graph[u].append(v)

    def dls(self, node, goal, depth):
        if depth == 0 and node == goal:
            return True
        if depth > 0:
            for neighbor in self.graph.get(node, []):
                if self.dls(neighbor, goal, depth - 1):
                    return True
        return False

    def iddfs(self, start, goal, max_depth):
        for depth in range(max_depth):
            if self.dls(start, goal, depth):
                return True
        return False

g = Graph()
g.add_edge('A', 'B')
g.add_edge('A', 'C')
g.add_edge('B', 'D')
g.add_edge('C', 'E')

start_node = 'A'
goal_node = 'D'
max_depth = 3

found = g.iddfs(start_node, goal_node, max_depth)
print(f"Goal {goal_node} {'found' if found else 'not found'} starting from {start_node}.")
```
