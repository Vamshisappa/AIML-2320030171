```python
from sklearn import datasets
import pandas as pd

diabetes = datasets.load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)

df['target'] = diabetes.target

print(df[['age', 'sex', 'bmi', 'target']].head())
```
