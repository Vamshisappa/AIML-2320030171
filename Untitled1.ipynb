```python
from collections import deque
def bfs(g, s):
  q=deque([s])
  v=set()
  v.add(s)
  while q:
    n=q.popleft()
    print(n,end=" ")
    for i in g[n]:
      if i not in v:
        v.add(i)
        q.append(i)
g={'A':['B','C'],
   'B':['D','E'],
   'C':['F'],
   'D':[],
   'E':[],
   'F':[]
   }
   bfs(g,'A')
```
